```python
import pandas as pd
import matplotlib.pyplot as plt
import json
from pandas.plotting import table

# Load JSON file into a DataFrame
def json_to_dataframe(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    df = pd.json_normalize(data)
    return df

# Create a UML diagram from the DataFrame
def create_uml_diagram(data_frame, diagram_title='UML Diagram'):
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.axis('off')
    table_data = ax.table(cellText=data_frame.values, colLabels=data_frame.columns, cellLoc='center', loc='center', colColours=['#f2f2f2']*len(data_frame.columns))
    table_data.auto_set_font_size(False)
    table_data.set_fontsize(10)
    table_data.scale(1.2, 1.2)
    ax.set_title(diagram_title, fontweight='bold', fontsize=14)
    plt.show()

# Example usage
json_file_path = 'JSON Files/duplexSmall.json'
df = json_to_dataframe(json_file_path)
create_uml_diagram(df)
```

```python
import json

# Step 1: Open JSON file
def load_json_file(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    return data

# Step 2: Update long URIs to short names
def update_uris(data):
    uri_mapping = {}  # Dictionary to store the mapping of original URIs to short names

    for item in data:
        for key, value in list(item.items()):
            if value.startswith('https://example.domain.de/'):
                original_uri = value
                if original_uri in uri_mapping:
                    short_name = uri_mapping[original_uri]
                else:
                    name = value.split('/')[-1].split('_')[0]
                    short_name = f'{name}#{len(uri_mapping) + 1}'
                    uri_mapping[original_uri] = short_name

                item[key] = short_name

    return data, uri_mapping

# Step 3: Save the URI mapping to a file
def save_uri_mapping(uri_mapping, mapping_file):
    with open(mapping_file, 'w') as file:
        json.dump(uri_mapping, file, indent=2)

# Step 4: Save the updated JSON file
def save_updated_json(data, output_file):
    with open(output_file, 'w') as file:
        json.dump(data, file, indent=2)
        
# Example usage
json_file_path = 'JSON Files/duplexSmall.json'
output_file_path = 'JSON Files/duplexSmallShort3.json'
uri_mapping_file_path = "JSON Files/uri_mapping1.txt"

json_data = load_json_file(json_file_path)
updated_data, uri_mapping = update_uris(json_data)
save_updated_json(updated_data, output_file_path)

# Save the URI mapping to a file
save_uri_mapping(uri_mapping, uri_mapping_file_path)
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

def process_json(json_data, hierarchy):
    unique_zones = set()
    unique_buildings = set()
    unique_storeys = set()
    unique_spaces = set()
    unique_elements = set()

    for item in json_data:
        unique_zones.add(item.get('zone'))
        unique_buildings.add(item.get('building'))
        unique_storeys.add(item.get('storey'))
        
        # Check if 'Space' key is present before adding to set
        if 'space' in item:
            unique_spaces.add(item.get('space'))
            
        unique_elements.add(item.get('element'))

    counts = [len(unique_zones), len(unique_buildings), len(unique_storeys), len(unique_spaces), len(unique_elements)]

    print(f'Counts for {hierarchy} Hierarchy:')
    print(f'Zones: {len(unique_zones)}')
    print(f'Buildings: {len(unique_buildings)}')
    print(f'Storeys: {len(unique_storeys)}')
    print(f'Spaces: {len(unique_spaces)}')
    print(f'Elements: {len(unique_elements)}\n')

    return counts

# Load JSON data from the Space hierarchy file
with open('JSON files\Storey-Space-Element-Hierarchy.json', 'r') as space_file:
    space_json_data = json.load(space_file)

# Load JSON data from the Storey hierarchy file
with open('JSON files\Storey-Element-Hierarchy.json', 'r') as storey_file:
    storey_json_data = json.load(storey_file)

# Process each hierarchy separately
space_counts = process_json(space_json_data, 'Space')
storey_counts = process_json(storey_json_data, 'Storey')

# Create a bar chart for each hierarchy
categories = ['Zones', 'Buildings', 'Storeys', 'Spaces', 'Elements']

plt.bar(categories, space_counts, color='blue', alpha=0.7, label='Zone-Building-Storey-Space-Element Hierarchy')
plt.bar(categories, storey_counts, color='orange', alpha=0.7, label='Zone-Building-Storey-Element Hierarchy', bottom=space_counts)

# Annotate each bar with the count
for i, count in enumerate(space_counts + storey_counts):
    plt.text(i, count + 0.1, str(count), ha='center')

# Set labels and title
plt.xlabel('Zones and Elements')
plt.ylabel('Count')
plt.title('Hierarchies')
plt.legend()

# Show the plot
plt.show()
```
